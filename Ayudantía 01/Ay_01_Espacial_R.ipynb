```r
## Librerías
library(ggplot2)
library(dplyr)
```

```r
## Parámetros
alpha <- 0.2   # nugget
beta  <- 0.8   # partial sill
range <- 1
```

```r
## Función del semivariograma esférico
gamma_spherical <- function(t, alpha, beta){

  ifelse(t == 0,
         0,
         ifelse(t < 1,
                alpha + beta * (3*t/2 - t^3/2),
                alpha + beta))
}

## Dominio
t_vals <- seq(0, 2, length.out = 500)

## Data frame
df <- data.frame(
  t = t_vals,
  gamma = gamma_spherical(t_vals, alpha, beta)
)
```

```r
## Esta línea es solo para agrandar el gráfico en colab
options(repr.plot.width = 12, repr.plot.height = 7)
```

```r
## Graficamos
ggplot(df, aes(t, gamma)) +
  geom_line(linewidth = 1.5, color = "#2C7FB8") +

  geom_hline(yintercept = alpha,
             linetype = "dashed",
             color = "#D7301F",
             linewidth = 1) +

  geom_hline(yintercept = alpha + beta,
             linetype = "dashed",
             color = "#1A9850",
             linewidth = 1) +

  geom_vline(xintercept = range,
             linetype = "dotted",
             linewidth = 1) +

  labs(
    title = "Semivariograma Esférico",
    subtitle = "Modelo utilizado en geoestadística",
    x = "Distancia |t|",
    y = expression(gamma(t))
  ) +

  annotate("text", x = 1.6, y = alpha, label = "Nugget", color = "#D7301F", size = 5) +
  annotate("text", x = 1.6, y = alpha + beta, label = "Sill", color = "#1A9850", size = 5) +
  annotate("text", x = range, y = 0.05, label = "Range", angle = 90, size = 5) +

  theme_minimal(base_size = 18) +

  theme(
    plot.title = element_text(face = "bold", size = 20),
    plot.subtitle = element_text(size = 16),
    panel.grid.minor = element_blank()
  )
```
